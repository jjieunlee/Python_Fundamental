```python
# 모든 정보를 담을 DF
movieIDDF=pd.DataFrame()

# 1개 영화 정보들에 대해서 처리할 DF
movieIDDF_temp=pd.DataFrame()
movieIDDF_temp=movieIDDF_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","openDt":"","directorpeopleNm":""},ignore_index=True)

num=len(movieIdData["movieListResult"]["movieList"])
for i in range(0,num):
    
    print(str(i)+"_번째 영화코드",movieIdData["movieListResult"]["movieList"][i]["movieCd"])
    #감독에 대한 정보가 뭐라도 들어있는 경우
    if movieIdData["movieListResult"]["movieList"][i]["directors"]!=[]:
        print(str(i)+"감독이름:",movieIdData["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"])
    else:
        print(str(i)+"감독이름: 정보 없음")
```

```python
# 모든 정보를 담을 DF

movieIDDF = pd.DataFrame()

# 1개 영화 정보들에 대해서 처리할 DF


num = len(movieIdData["movieListResult"]["movieList"])
for i in range(0, num):
    movieIDDF_temp = pd.DataFrame()
    movieIDDF_temp = movieIDDF_temp.append({"movieCd" : "","movieNm" : "","movieNmEn" : "","openDt" : "","directorpeopleNm" : ""}, ignore_index=True)

    # 1.moiveIDDF_temp에 i번째 영화의 정보들을 담기..(코드, 이름, 영문, 개봉일, 감독)
    # 2.movieIDDF_temp에 있는 i번째 영화에 대한 정보를 movieIDDF에 추가!!!
    # -------> append
    # 3.movieIDDF_temp에 담겨져 있는 정보들을 초기화 하는 부분...
    movieIDDF_temp["movieCd"] = movieIdData["movieListResult"]["movieList"][i]["movieCd"]
    movieIDDF_temp["movieNm"] = movieIdData["movieListResult"]["movieList"][i]["movieNm"]
    movieIDDF_temp["movieNmEn"] = movieIdData["movieListResult"]["movieList"][i]["movieNmEn"]
    movieIDDF_temp["openDt"] = movieIdData["movieListResult"]["movieList"][i]["openDt"]
    # movieIDDF_temp["directorpeopleNm"] = movieIdData["movieListResult"]["movieList"][i]["directorpeopleNm"]
    if movieIdData["movieListResult"]["movieList"][i]["directors"] ==[]:
        movieIDDF_temp["directorpeopleNm"] = " "
    else:
        movieIDDF_temp["directorpeopleNm"] = movieIdData["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"]
   
    
    movieIDDF = movieIDDF.append(movieIDDF_temp, ignore_index=True)
movieIDDF
```
